```python
import numpy as np
import pandas as pd
np.random.seed(25)
data={
    "x" : np.arange(1,10),
    "y" : np.random.randint(1, 99, size=9)
}
df=pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (5,5))
plt.plot(df["x"],df["y"],marker='o',color='blue')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Random Numbers")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["A", "B", "C", "D"]
y = [40, 70, 30, 90]
plt.figure(figsize = (6,4))
plt.bar(x, y, color='green')
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales Bar Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title("Line")
plt.subplot(2, 2, 2)
plt.bar(x, y)
plt.title("Bar")
plt.subplot(2, 2, 3)
plt.scatter(x, y)
plt.title("Scatter")
plt.subplot(2, 2, 4)
plt.pie(y, labels=x)
plt.title("Pie")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(200)
plt.figure(figsize=(6,4))
plt.hist(data, bins=20, color='green', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of 200 Random Normal Values")
plt.show()
```
